```
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y=np.sin(x)

plt.plot(x,y,label="Sine Wave")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Line Plot")
plt.legend()
plt.show()
```

```
categories=['A','B','C','D']
values=[10,25,25,30]

plt.figure(figsize=(6,4))
plt.bar(categories,values,color='purple')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title(" Bar  chart Example")
plt.show()
```

```
data=np.random.randn(1000)

plt.figure(figsize=(7,5))
plt.hist(data,bins=30,color='red',edgecolor='black',alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```
import seaborn as sns
import pandas as pd


data = np.random.randn(1000)
df = pd.DataFrame(data,columns=['Values'])

sns.histplot(df['Values'],bins=30,kde=True,color='blue')
plt.title("Histogram with KDE")
plt.show()
```

```
tips = sns.load_dataset('tips')

plt.figure(figsize=(6,4))
sns.boxplot(x=tips['total_bill'])
plt.title ("Box  plot pf Total Bill")
plt.show()
```

```
corr_matrix = tips[['total_bill', 'tip', 'size']].corr()
plt.figure(figsize=(7,5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```
import plotly.express as px
import numpy as np
import pandas as pd

# Create data
x = np.linspace(0, 10, 100)  # Generate 100 points between 0 and 10
y = np.sin(x)  # Compute sine values

# Create DataFrame
df = pd.DataFrame({"x": x, "y": y})

# Create interactive line plot
fig = px.line(df, x="x", y="y", title="Interactive Sine Wave")

# Show the figure
fig.show()
```

```
import plotly.express as px

# Load sample dataset
df = px.data.tips()

# Create interactive scatter plot
fig = px.scatter(df,
                 x="total_bill",
                 y="tip",
                 color="sex",
                 size="size",
                 title="Total Bill vs Tip")

# Show the figure
fig.show()
```

```
import plotly.graph_objects as go
import plotly.express as px

# Load dataset
df = px.data.tips()

# Create 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=df["total_bill"],
    y=df["tip"],
    z=df["size"],
    mode="markers",
    marker=dict(
        size=5,
        color=df["total_bill"],  # Color by total bill
        colorscale="Viridis"  # Color scheme
    )
)])

# Update layout
fig.update_layout(title="3D Scatter Plot of Total Bill, Tip & Size",
                  scene=dict(xaxis_title="Total Bill",
                             yaxis_title="Tip",
                             zaxis_title="Size"))

# Show the figure
fig.show()
```
